```python
#Importar librerias necesarias
import funciones as fc
import pandas as pd

# Ajustar la configuración de visualización
pd.set_option('display.max_columns', None)  
```

```python
# Lista de tablas a procesar
tablas = ['HAPMEDIDA','APMEDIDA_CO','HFACTURACION','ORDENES','RECIBOS','TARIFAS','TRABPEND_FR','TRABPEND_RE']
```

```python
#Almacenar los dataframes en diccionarios y validar tamaño de cada dataframe

diccionario_dataframes = {}

for tabla in tablas:
    df=fc.conexion_bd(tabla=tabla)
    diccionario_dataframes[tabla] = df
```

```python
diccionario_dataframes_copia=diccionario_dataframes
```

```python
#Conocer la cantidad de registros por tipo de columna en cada Dataframe

for tabla,df in diccionario_dataframes.items():
    print("="*20)
    value_counts_result = df.dtypes.value_counts()
    print(f"Cantidad de columnas {tabla} por tipo:")
    print(value_counts_result)
    
```

```python
# Calcular la cantidad de valores nulos por columna en cada Dataframe

for tabla,df in diccionario_dataframes.items():
    
    print("="*20)

    null_counts = df.isnull().sum()

    # Filtrar columnas con valores nulos
    null_counts = null_counts[null_counts >1]

    # Ordenar de mayor a menor cantidad de valores nulos
    null_counts = null_counts.sort_values(ascending=True)

    print(f"Analisis de valores nulos {tabla}: \n{null_counts}")
```

```python
# Analisis de valores unicos en las columnas de tipo texto para cada Dataframe

for tabla, df in diccionario_dataframes.items():

    print("Analisis valores unicos columnas tipo texto")

    total_filas = len(df)

    columnas_texto = df.select_dtypes(include='object').columns

    columna_nulas = null_counts[null_counts/total_filas > 0.59].index.tolist()

    for columna in columnas_texto:
        
        if df[columna].value_counts().mean()>1 and columna not in columna_nulas:
            
            print("="*100)

            porcentaje_valores_unicos = round(len(df[columna].value_counts())*100/len(df),2)
            
            print(f"Valores unicos columna {columna} | % valores unicos ({porcentaje_valores_unicos}):")
            
            print(df[columna].value_counts())
```

```python
# Depurar los dataframes con la función clean_data

for tabla,df in diccionario_dataframes.items():
    
    fc.clean_data(df)
    print(f'Data depurada {tabla}')
```

```python
## Normalizar valores de las filas

columnas =['USUARIO','PROGRAMA']

for nombre_df, df in diccionario_dataframes.items():
    for columna in columnas:
        if columna in df.columns:
            df[columna] = df[columna].str.lower().str.strip()
            


```

```python
# Analisis de valores unicos en las columnas de tipo texto para cada Dataframe

for tabla, df in diccionario_dataframes.items():

    print(f"Analisis valores unicos columnas tipo texto en dataframe {tabla}")

    total_filas = len(df)

    columnas_texto = df.select_dtypes(include='object').columns

    columna_nulas = null_counts[null_counts/total_filas > 0.59].index.tolist()

    for columna in columnas_texto:
        
        if df[columna].value_counts().mean()>1 and columna not in columna_nulas:
            
            print("="*100)

            porcentaje_valores_unicos = round(len(df[columna].value_counts())*100/len(df),2)
            
            print(f"Valores unicos columna {columna} | % valores unicos ({porcentaje_valores_unicos}):")
            
            print(df[columna].value_counts())
```
